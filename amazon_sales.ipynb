### bought_in_last_month

```python
df["bought_in_last_month"].head()
```

```python
df["bought_in_last_month"].unique()
```

```python
import numpy as np

def clean_bought(x):
    if pd.isna(x):
        return np.nan

    x = str(x)

    if 'bought in past month' in x:
        num = x.split('+')[0]

        if 'K' in num:
            return float(num.replace('K', '')) * 1000

        return float(num)

    return np.nan

df['bought_in_last_month'] = df['bought_in_last_month'].apply(clean_bought)
```

```python
df['bought_in_last_month'].isnull().sum()
```

```python
df['bought_in_last_month'].fillna(
    df['bought_in_last_month'].median(),
    inplace=True
)
```

```python
df['bought_in_last_month'].isnull().sum()
```

```python
df.isnull().sum()
```
